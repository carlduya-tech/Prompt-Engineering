```python
#!/usr/bin/env python3
"""
Simple "Airbnb-style" navigator for a curated list of stays.
- Browse stays
- Filter by max price, min bedrooms, min beds
- Sort results by price

Run:
  python stays_navigator.py
"""

from dataclasses import dataclass
from typing import List, Optional


@dataclass(frozen=True)
class Stay:
    id: int
    name: str
    area: str
    price_per_night: int  # USD
    bedrooms: int
    beds: int
    sleeps: int
    notes: str


STAYS: List[Stay] = [
    Stay(1, "Anaheim Resort Area Family Home", "Anaheim (Resort Area)", 475, 3, 4, 8,
         "Closest-commute style stay; great for rope-drop mornings."),
    Stay(2, "West Anaheim Value Townhome", "Anaheim (West)", 295, 2, 3, 6,
         "Value pick with easy parking and quick freeway access."),
    Stay(3, "Anaheim Hills Spacious Retreat", "Anaheim Hills", 595, 4, 6, 10,
         "Bigger home; quieter neighborhood; more space for groups."),
    Stay(4, "Garden Grove Family Home w/ Yard", "Garden Grove", 365, 3, 4, 8,
         "Residential feel; typically good value near the Anaheim border."),
    Stay(5, "Fullerton Craftsman-Style Home", "Fullerton", 410, 3, 4, 8,
         "Charming area; good dining; solid family option."),
    Stay(6, "Buena Park Park-Hopper Home", "Buena Park", 330, 3, 4, 8,
         "Great if mixing Disney + Knott’s style days."),
    Stay(7, "Old Towne Orange Updated Bungalow", "Orange", 345, 2, 3, 6,
         "Smaller footprint; walkable vibes; easy in/out."),
    Stay(8, "Long Beach Coastal Home", "Long Beach", 485, 3, 4, 8,
         "Beach + Disney balance; commute tradeoff for ocean proximity."),
    Stay(9, "Downey Family Home (LA County)", "Downey", 320, 3, 4, 8,
         "Good compromise for LA County + Disney access."),
    Stay(10, "Pasadena Bungalow", "Pasadena", 440, 2, 3, 6,
         "Best for LA attractions first, Disney as a day trip."),
]


def apply_filters(
    stays: List[Stay],
    max_price: Optional[int],
    min_bedrooms: Optional[int],
    min_beds: Optional[int],
) -> List[Stay]:
    results = stays
    if max_price is not None:
        results = [s for s in results if s.price_per_night <= max_price]
    if min_bedrooms is not None:
        results = [s for s in results if s.bedrooms >= min_bedrooms]
    if min_beds is not None:
        results = [s for s in results if s.beds >= min_beds]
    return results


def print_stays(stays: List[Stay]) -> None:
    if not stays:
        print("\nNo stays match your current filters.\n")
        return

    print("\nResults:")
    print("-" * 78)
    for s in stays:
        print(f"[{s.id:02d}] ${s.price_per_night}/night | {s.bedrooms} bd | {s.beds} beds | "
              f"sleeps {s.sleeps} | {s.area}")
        print(f"     {s.name}")
    print("-" * 78)


def print_stay_detail(stay: Stay) -> None:
    print("\n" + "=" * 78)
    print(f"{stay.name} (ID {stay.id})")
    print("=" * 78)
    print(f"Area:          {stay.area}")
    print(f"Price/night:   ${stay.price_per_night}")
    print(f"Bedrooms:      {stay.bedrooms}")
    print(f"Beds:          {stay.beds}")
    print(f"Sleeps:        {stay.sleeps}")
    print(f"Notes:         {stay.notes}")
    print("=" * 78 + "\n")


def prompt_int(label: str) -> Optional[int]:
    raw = input(label).strip()
    if raw == "":
        return None
    try:
        value = int(raw)
        if value < 0:
            print("Please enter a non-negative number (or press Enter to skip).")
            return prompt_int(label)
        return value
    except ValueError:
        print("Please enter a whole number (or press Enter to skip).")
        return prompt_int(label)


def main() -> None:
    max_price: Optional[int] = None
    min_bedrooms: Optional[int] = None
    min_beds: Optional[int] = None
    sort_by_price: bool = True

    while True:
        filtered = apply_filters(STAYS, max_price, min_bedrooms, min_beds)
        if sort_by_price:
            filtered = sorted(filtered, key=lambda s: s.price_per_night)

        print_stays(filtered)

        print("Menu:")
        print("  1) Set / update filters")
        print("  2) Clear all filters")
        print("  3) Toggle sort by price (currently: {})".format("ON" if sort_by_price else "OFF"))
        print("  4) View a stay in detail (by ID)")
        print("  5) Exit")

        choice = input("\nChoose an option (1-5): ").strip()

        if choice == "1":
            print("\nEnter filters (press Enter to skip any filter):")
            max_price = prompt_int("  Max price per night (USD): ")
            min_bedrooms = prompt_int("  Minimum bedrooms: ")
            min_beds = prompt_int("  Minimum beds: ")
        elif choice == "2":
            max_price = None
            min_bedrooms = None
            min_beds = None
            print("\nFilters cleared.\n")
        elif choice == "3":
            sort_by_price = not sort_by_price
        elif choice == "4":
            stay_id = prompt_int("Enter stay ID: ")
            if stay_id is None:
                continue
            match = next((s for s in STAYS if s.id == stay_id), None)
            if match:
                print_stay_detail(match)
            else:
                print("No stay found with that ID.\n")
        elif choice == "5":
            print("Goodbye!")
            return
        else:
            print("Please choose a valid option (1-5).\n")


if __name__ == "__main__":
    main()
```
